```python
from spherical_analysis import *
import os
import json
```

This example notebook will demonstrate how to visualize the centers and radii that are detected by the center finding script. Useful for validation. Assumes you have a `.json` file with centers and radii.

```python
# paremters
json_dir = r"/mnt/data1/Code/GH_local/spherical_harmonic_rotation/dataset_5/"
json_name = r"dataset_5.json"

# reads in json 
with open(os.path.join(json_dir, json_name), 'r') as f:
    dataset = json.load(f)

# sort by time if not already
dataset = dict(sorted(dataset.items(), key=lambda x: int(x[0][4:])))

# convert back necessary values back to nd_array
for d in dataset.values():
    d['center'] = np.array(d['center'])
```

Visualize the center for a single time point. We'll do that with the function `visualize_center`

```python
# read in a timepoint
d = dataset['time90']
im = clip_pad_image(io.imread(d['path']), d['alpha'], d['padding'])

# plot
visualize_center(im, d['center'], d['radius'])
```

Here is an example of doing this with all points in the dataset, and saving the results in a folder.

```python
# make directory for figures
save_folder = r"/mnt/data1/Code/GH_local/spherical_harmonic_rotation_analysis/DS5_centers_radii"
try:
    os.mkdir(save_folder)
except:
    print("Directory already exists")
```

```python
# saves a plot of the image with center and circumcircle
for token in dataset.keys():
    print(f"Processing {token}")
    d = dataset[token]
    im = clip_pad_image(io.imread(d['path']), d['alpha'], d['padding'])
    visualize_center(im, d['center'], d['radius'])
    plt.savefig(os.path.join(save_folder, f'{token}_center_radii.png'), format='png', dpi=300)
    plt.close('all')
```
